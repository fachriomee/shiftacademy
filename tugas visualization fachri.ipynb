```python
#film yg muncul di netflix setiap tahunnya dari data1
Netflix = data1.groupby('Year')['Netflix'].sum().reset_index()
Netflix.plot(x='Year', y='Netflix', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#film yg muncul di netflix setiap tahunnya dari data2
Netflix2 = data2.groupby('Year')['Netflix'].sum().reset_index()
Netflix2.plot(x='Year', y='Netflix', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#film yg muncul di prime video setiap tahunnya dari data1
Prime_Video = data1.groupby('Year')['Prime Video'].sum().reset_index()
Prime_Video.plot(x='Year', y='Prime Video', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#film yg muncul di prime video setiap tahunnya dari data2
Prime_Video2 = data2.groupby('Year')['Prime Video'].sum().reset_index()
Prime_Video2.plot(x='Year', y='Prime Video', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#film yg muncul di disney setiap tahunnya dari data1
Disney= data1.groupby('Year')['Disney+'].sum().reset_index()
Disney.plot(x='Year', y='Disney+', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#film yg muncul di disney setiap tahunnya dari data2
Disney2= data2.groupby('Year')['Disney+'].sum().reset_index()
Disney2.plot(x='Year', y='Disney+', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#film yg muncul di hulu setiap tahunnya dari data1
Hulu= data1.groupby('Year')['Hulu'].sum().reset_index()
Hulu.plot(x='Year', y='Hulu', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#film yg muncul di hulu setiap tahunnya dari data2
Hulu2= data2.groupby('Year')['Hulu'].sum().reset_index()
Hulu2.plot(x='Year', y='Hulu', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#perbandingan disney, netflix, prime video, dan hulu dari dari data1
ax= sns.lineplot(x='Year', y='Netflix', data=Netflix, color='red')
ay=sns.lineplot(x='Year', y='Prime Video', data=Prime_Video, color='blue')
az=sns.lineplot(x='Year', y='Hulu', data=Hulu, color='yellow')
aq=sns.lineplot(x='Year', y='Disney+',data=Disney,color='green' )
plt.legend(labels=["Netflix","Prime Video", "Hulu", "Disney+"])
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#perbandingan disney, netflix, prime video, dan hulu dari dari data2
ax= sns.lineplot(x='Year', y='Netflix', data=Netflix2,color='red')
ay=sns.lineplot(x='Year', y='Prime Video', data=Prime_Video2,color='blue')
az=sns.lineplot(x='Year', y='Hulu', data=Hulu2, color='yellow')
aq=sns.lineplot(x='Year', y='Disney+',data=Disney2, color='green')
plt.legend(labels=["Netflix","Prime Video", "Hulu", "Disney+"])
plt.xlabel("Year")
plt.ylabel("Sum Of Film")
```

```python
#untuk mengetahui rata-rata run time/durasi pemutaran film dari masing masing negara
data13 = data1.groupby('Country')['Runtime'].mean().reset_index()
data13.plot(x='Country', y='Runtime', kind='line', figsize=(10, 18))
plt.xlabel("Country")
plt.ylabel("Sum Of Runtime")
```

```python
#untuk mengetahui run time/durasi pemutaran film terbanyak dari 5 Negara dari data1
data14= data1.groupby('Country')['Runtime'].sum().reset_index()
data14.sort_values(by='Runtime', ascending =False, inplace = True)
a= data14.head()
pd.DataFrame(a).plot(x='Country', y='Runtime', kind='bar')
plt.xlabel('Country')
plt.ylabel('Sum Of Runtime')
```

```python
#mengetahui rata-rata durasi film di setiap kategori usia dari data1
data11 = data1.groupby('Age')['Runtime'].mean().reset_index()
data11.plot(x='Age', y='Runtime', kind='bar')
plt.xlabel("Age")
plt.ylabel("Sum Of Runtime")
```

```python
#jumlah menit film baru yang muncul setiap tahunnya dari data1
data12 = data1.groupby('Year')['Runtime'].sum().reset_index()
data12.plot(x='Year', y='Runtime', kind='line')
plt.xlabel("Year")
plt.ylabel("Sum Of Runtime")
```

```python
#partisipasi setiap bahasa dalam pembuatan setiap film untuk setiap tahunnya dari data1
bahasa = data1.groupby('Year')['Language'].sum().reset_index()
pd.DataFrame(bahasa).plot()
plt.xlabel("Language")
plt.ylabel("Year")
```

```python
#partisipasi setiap negara dalam pembuatan setiap film untuk setiap tahunnya dari data1
Negara = data1.groupby('Year')['Country'].sum().reset_index()
pd.DataFrame(Negara).plot()
plt.xlabel("Country")
plt.ylabel("Year")
```

```python
#untuk mengetahui run time/durasi pemutaran film terbanyak dari 5 direktur data1
data22= data1.groupby('Directors')['Runtime'].sum().reset_index()
data22.sort_values(by='Runtime', ascending =False, inplace = True)
p= data22.head()
pd.DataFrame(p).plot(x='Directors', y='Runtime', kind='bar')
plt.xlabel('Directors')
plt.ylabel('Sum Of Runtime')
```
